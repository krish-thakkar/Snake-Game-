```python
#6  Python Program to Solve Quadratic Equations. 
import cmath

def solve_quadratic(a, b, c):
    # Calculate the discriminant
    discriminant = (b ** 2) - (4 * a * c)

    # Compute the solutions
    solution1 = (-b - cmath.sqrt(discriminant)) / (2 * a)
    solution2 = (-b + cmath.sqrt(discriminant)) / (2 * a)

    return solution1, solution2

# Get coefficients from the user
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

# Solve the quadratic equation
solution1, solution2 = solve_quadratic(a, b, c)

# Display the solutions
print(f"Solutions for the quadratic equation {a}x^2 + {b}x + {c} = 0 are:")
print(f"x1 = {solution1}")
print(f"x2 = {solution2}")
```

```python
#7 Write a program to print a specified number of Fibonacci numbers.
def generate_fibonacci(count):
    fibonacci_sequence = []
    a, b = 0, 1

    for _ in range(count):
        fibonacci_sequence.append(a)
        a, b = b, a + b

    return fibonacci_sequence

# Get the number of Fibonacci numbers to generate
num = int(input("Enter the number of Fibonacci numbers to generate: "))

# Generate and print the Fibonacci sequence
fibonacci_numbers = generate_fibonacci(num)
print(f"Fibonacci sequence with {num} numbers:")
print(fibonacci_numbers)
```

```python
#14 Write a function to calculate the arithmetic mean of a variable number of values.
def calculate_mean(*values):
    if len(values) == 0:
        return 0  # Return 0 if no values are provided to avoid division by zero

    total = sum(values)
    mean = total / len(values)
    return mean

# Example usage:
values = [4, 7, 10, 3, 8]

# Calculate the mean of the values using the function
result = calculate_mean(*values)

print(f"The mean of the values is: {result}")
```

```python
#16 Write a program to find the factorial of a given number using recursion.
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Get the number from the user
num = int(input("Enter a number to find its factorial: "))

# Calculate factorial using the factorial() function
result = factorial(num)

print(f"The factorial of {num} is: {result}")
```

```python
"""
18 Implement a simple banking application with methods: 
create Account(),
 deposit()
, withdraw(),
 computeInterest(),
displayBalance()
"""
class BankAccount:
    def __init__(self, account_number, initial_balance=0, interest_rate=0.05):
        self.account_number = account_number
        self.balance = initial_balance
        self.interest_rate = interest_rate

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}")
        else:
            print("Invalid deposit amount. Please deposit a positive value.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}")
        else:
            print("Insufficient funds or invalid amount for withdrawal.")

    def compute_interest(self):
        interest_amount = self.balance * self.interest_rate
        self.balance += interest_amount
        print(f"Interest of ${interest_amount} computed and added to the balance.")

    def display_balance(self):
        print(f"Account Number: {self.account_number}")
        print(f"Current Balance: ${self.balance}")


# Taking user input
account_number = input("Enter account number: ")
initial_balance = float(input("Enter initial balance: "))
interest_rate = float(input("Enter interest rate (in decimal): "))

account1 = BankAccount(account_number, initial_balance, interest_rate)

while True:
    print("\n1. Deposit\n2. Withdraw\n3. Compute Interest\n4. Display Balance\n5. Exit")
    choice = input("Enter choice (1-5): ")

    if choice == '1':
        amount = float(input("Enter deposit amount: "))
        account1.deposit(amount)
    elif choice == '2':
        amount = float(input("Enter withdrawal amount: "))
        account1.withdraw(amount)
    elif choice == '3':
        account1.compute_interest()
    elif choice == '4':
        account1.display_balance()
    elif choice == '5':
        print("Exiting the banking application. Thank you!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```

```python
#19 Create a user-defined exception to display a proper message when the value of (b*d) is zero.
class ZeroProductError(Exception):
    def __init__(self, message="The product of b and d is zero."):
        self.message = message
        super().__init__(self.message)
def calculate_product(b, d):
    product = b * d
    if product == 0:
        raise ZeroProductError()
    return product

# Example usage:
b_value = 0
d_value = 10

try:
    result = calculate_product(b_value, d_value)
    print(f"The product of b and d is: {result}")
except ZeroProductError as e:
    print(f"Error: {e}")
```

```python
import csv
import random

# Generate some sample data
data = []
for _ in range(100):
    age = random.randint(20, 60)
    job = random.choice(['admin.', 'blue-collar', 'technician', 'services', 'management'])
    marital = random.choice(['married', 'single', 'divorced'])
    education = random.choice(['primary', 'secondary', 'tertiary'])
    balance = random.randint(-1000, 5000)
    subscribed = random.choice(['yes', 'no'])

    data.append([age, job, marital, education, balance, subscribed])

# Write data to CSV file
with open('bank_marketing.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(['age', 'job', 'marital', 'education', 'balance', 'y'])
    # Write data rows
    writer.writerows(data)
```

```python
#30 Develop a script to find the GCD (Greatest Common Divisor) of two numbers.
def gcd(x,y):
    if (y==0):
        return abs(x)
    else:
        return gcd(y,x%y)
x = int(input("enter the value of x:"))
y = int(input("enter the value of y:"))
result = gcd(x,y)
print(result)
```

```python
#31 Write a Python function to find the sum of digits of a number using recursion.
def sum_of_digits(n):
    if n < 10:
        return n
    else:
        return n % 10 + sum_of_digits(n // 10)

# Example usage:
number = int(input("Enter a number: "))
result = sum_of_digits(number)
print(f"The sum of digits of {number} is: {result}")
```
